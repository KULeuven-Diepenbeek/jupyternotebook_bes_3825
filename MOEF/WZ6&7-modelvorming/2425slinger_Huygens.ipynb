### Analyse slinger van Huygens

```python
import matplotlib.pyplot as plt
%matplotlib widget
import numpy as np
import ipywidgets
```

```python
T = 4.00     # (s) meting van de 1 periode van de slinger
dT = 0.01     # (s) absolute onzekerheid op de meting van de 1 periode van de slinger
g = 9.81     # (m/sÂ²) valversnelling
dg = 0.01    # (m/sÂ²) absolute onzekerheid op de valversnelling

def berekenAantalPeriodes(dL):
    return 2*g*T*dT/(4*np.pi**2*dL-T**2*dg)

# figuur 1
dntot=np.linspace(0.1,1,250)
xAs = list(dntot)
yAs = list(map(berekenAantalPeriodes, xAs))
plt.figure()
plt.plot(xAs, yAs, color = 'b')
plt.grid(True)
plt.xlabel("absolute onzekerheid op de lengte van het touw (m)")
plt.ylabel("Aantal te meten periodes (-)")
plt.show()


```

```python
def plotVarieerT(t1):
    nmax = 20
    n=np.linspace(0.1,nmax,250)
    def berekendLifvdT(n1):
        return t1**2*dg/(4*np.pi**2)+g*t1*dT/(2*np.pi**2*n1)
    xAs = list(n)
    yAs = list(map(berekendLifvdT, xAs))
    plt.figure()
    plt.plot(xAs, yAs, color = 'b')
    plt.grid(True)
    plt.xlabel("Aantal gemeten periodes (-)")
    plt.ylabel("Absolute onzekerheid op de lengte van het touw (m)")
    plt.ylim(0 , 0.4)
    #plt.yscale("log")
    plt.show()

sliderT = ipywidgets.IntSlider(value=5, min=0.5, max=100, step = 1, description = "periode (s)")
grafiek = ipywidgets.interactive(plotVarieerT, t1=sliderT)
display(grafiek)
```

```python
T = 4.00     # (s) meting van de 1 periode van de slinger
dT = 0.01     # (s) absolute onzekerheid op de meting van de 1 periode van de slinger
g = 9.81     # (m/sÂ²) valversnelling
dg = 0.01    # (m/sÂ²) absolute onzekerheid op de valversnelling
xAsym=T**2*dg/4/np.pi**2

 
def plotAantalPeriodes(dLMin, dLMax):
    dL=np.linspace(dLMin,dLMax,250)
    xAs = list(dL)
    yAs = list(map(berekenAantalPeriodes, xAs))
    plt.figure()
    plt.plot(xAs, yAs, color = 'b')
    plt.grid(True)
    plt.xlabel("Gewenste absolute onzekerheid op de lengte (m)")
    plt.ylabel("Aantal periodes te meten (-)")
    #plt.xscale("log")
    #plt.yscale("log")
    plt.show()

sliderdLMin = ipywidgets.FloatSlider(value=0.5, min=xAsym, max=0.8, step = 0.05, description = "dLMin")
sliderdLMax = ipywidgets.FloatSlider(value=5, min=1, max=10, step = 0.20, description = "dLMax")
grafiek = ipywidgets.interactive(plotAantalPeriodes, dLMin=sliderdLMin,dLMax=sliderdLMax)
display(grafiek)
```
