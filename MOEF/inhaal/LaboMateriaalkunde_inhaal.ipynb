## Achtergrondinformatie
De **spanning** in een materiaal is $\sigma = \frac{F}{A}$ met $F$ de normaalkracht en $A$ de oppervlakte van de doorsnede.

De **rek** is gegeven door $\varepsilon = \frac{\Delta L}{L}$ met $\Delta L$ de uitrekking en $l$ de lengte van de staaf.

De **$E$-modulus** is het verband tussen spanning en rek: $\sigma = E \varepsilon$
